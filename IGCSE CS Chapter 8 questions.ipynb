```python
#Question 11

age = int(input("Input your Age: "))
years_lived = int(input("Input how many Years lied in town: "))
parents_in_town = int(input("Number of your parents living in this town: "))
parent_passed = input("Is one of your parent deceased? (y/n): ")
sibling_years = int(input("How many Years has your sibling( lived n town: "))
money = float(input("what is your Bnk balance: "))
prison = input("Have you Ever been to prison? (y/n): ")
support = int(input("Number of memmmbers supporting you: "))

residency_pass = False

if years_lived >= 20:
    residency_pass = True
elif years_lived >= 5 and parents_in_town == 2:
    residency_pass = True
elif parents_in_town == 1 and parent_passed == "y":
    residency_pass = True
elif sibling_years >= 10:
    residency_pass = True

errors = ""
error_count = 0
age_failed = False

if age < 18 or age > 65:
    errors += "- age not meeting requirements "
    error_count += 1
    age_failed = True

if residency_pass == False:
    errors += "- Residency requirement not met"
    error_count += 1

if money < 500000:
    errors += "- Not enough money"
    error_count += 1

if prison == "y":
    errors += "- You have a prison record"
    error_count += 1

if support < 2:
    errors += "- Not enough support (need at least 2)"
    error_count += 1

if error_count == 0:
    print("Accepted! Your ID is: 7721")
elif support >= 5 and error_count == 1 and age_failed == False:
    print("Accepted! Your ID is: 7721")
else:
    print("Application Rejected for the following reasons:")
    print(errors)
```
